```python
def few_shots(n=10, balanced=True):
    if not balanced:
        return dev_data.sample(n)
    rst = pd.concat([dev_data[dev_data['relation_variable_label'] == 1].sample(n//2), dev_data[dev_data['relation_variable_label'] == 0].sample(n - n//2)])
    rst = rst.sample(frac=1)
    return rst[['label_1_processed', 'label_2_processed', 'relation_variable_label', 'label_1_table_path', 'label_2_table_path', 'label_1_col_idx', 'label_2_col_idx']]
```

```python
def sample_column_value(table_path, col_id, n=5):
    table = pd.read_json(table_path, compression="gzip", lines=True)
    table = table.astype(str)
    k = n
    rst = table[col_id].unique().tolist()[:k]
    while len("".join(rst)) > 300:
        k -= 1
        rst = table[col_id].unique().tolist()[:k]
    if k:
        return table[col_id].unique().tolist()[:k]
    else:
        return table[col_id].unique().tolist()[:1]
```

```python
prompts = []
labels = []

for _, row in test_data.iterrows():
    prompt = """Column semantic types are used to describe the semantic type of a table column. Different systems use sementic type labels from different ontologies, your task is to determine whether two sementic types are equivelant. For each input label, you will also be provided with sample column values from columns labeled with the input label. Return your final result in the following JSON format : {"answer": <yes or no>}.

For example:

"""
    fewshots_prompt = "\n\n".join(["Input:\nLabel 1: {}\nSample Column Value for Label 1:\n{}\n\nLabel 2: {}\nSample Column Value for Label 2:\n{}\n\nOutput:\n{}".format(row[0], sample_column_value(row[3], row[5]), row[1], sample_column_value(row[4], row[6]), '{"answer": "yes"}' if row[2] else '{"answer": "no"}') for row in few_shots().values.tolist()])
    prompt += fewshots_prompt
    prompt += """

Now, for the following label pairs, please determine if they are equivelant. Return your final result in the following JSON format : {{"answer": <yes or no>}}.

Input:
Label 1: {}
Sample Column Value for Label 1:
{}

Label 2: {}
Sample Column Value for Label 2:
{}

Output:
    """.format(row['label_1_processed'], sample_column_value(row['label_1_table_path'], row['label_1_col_idx']), row['label_2_processed'], sample_column_value(row['label_2_table_path'], row['label_2_col_idx']))
    prompts.append(prompt)
    labels.append(row['relation_variable_label'])
```
